```python
#Critter Cam
# import GoPiGo Modules:
from easygopigo3 import EasyGoPiGo3

#import other stuff:
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
from EDL_Jupyter_resources import HiddenPrints
hiddenprints=HiddenPrints()

# Initialize easy gpg
easyGPG = EasyGoPiGo3()
# Make sure that all sensors and output devices are uncofigured to start
easyGPG.reset_all()

#initialize the loudness sensor
US_sensor=easyGPG.init_ultrasonic_sensor('AD1')
loudness_sensor=easyGPG.init_loudness_sensor('AD2')

###Initialization of variables###
mean_distance=0
mean_loudness=0
recent_distance_data=np.array([])
recent_loudness_data=np.array([])
points=[]
all_time_distance=np.array([])  #debugging
all_time_loudness=np.array([])  #debugging
means_distance=np.array([])  #debugging
means_loudness=np.array([])  #debugging
times=np.array([])  #debugging

###Constants###
distance_thresh=0.5 #Dist must be smaller than this to execute
loudness_thresh=1.1 #Loudness must be larger than this to execute
history_length=5 #How long to keep data in seconds
sample_rate=30 #Sample rate for both sensors in Hz
length=20 #How long to record for in seconds

###Loop###
print('Starting')
start=time.time()
while time.time()-start<length:
    
    with hiddenprints: #suppress console output
        distance_reading=US_sensor.read() 
        loudness_reading=loudness_sensor.read() #get the reading from the sensors
    
    recent_distance_data=np.append(recent_distance_data,distance_reading)
    recent_loudness_data=np.append(recent_loudness_data,loudness_reading) #add the readings to our recent data
    all_time_distance=np.append(all_time_distance,distance_reading) #debugging
    all_time_loudness=np.append(all_time_loudness,loudness_reading) #debugging
    
    if len(recent_distance_data)+1>(history_length*sample_rate): #trim the readings if they are long enough
        recent_distance_data=np.delete(recent_distance_data,0)
    if len(recent_loudness_data)+1>(history_length*sample_rate):
        recent_loudness_data=np.delete(recent_loudness_data,0)
      
    if len(recent_distance_data)==0:  #calculate the means if the data is long enough
        mean_distance=distance_reading
    else:
        mean_distance=np.mean(recent_distance_data)
         
    if len(recent_loudness_data)==0: #for loudness too
        mean_loudness=loudness_reading
    else:
        mean_loudness=np.mean(recent_loudness_data)
    
    if loudness_reading>(mean_loudness*loudness_thresh) and (distance_reading<(mean_distance*distance_thresh)):
        print('Take a photo!') #if the loudness, and distance, are above/below thresh then take a photo
        points.append([time.time()-start,distance_reading,loudness_reading]) #record the point
    
    means_distance=np.append(means_distance,mean_distance) #debugging
    means_loudness=np.append(means_loudness,mean_loudness) #debugging
    times=np.append(times,time.time()-start) #debugging
    
    time.sleep(1.0/sample_rate) #sleep a bit
    
print("Done!")
print('Found unusual movement at:')
for i in range(len(points)):
    print('T+: %f'%(points[i][0]))

'''Debugging'''
plt.figure(0)
plt.plot(times,all_time_loudness)
plt.plot(times,means_loudness)
for i in range(len(points)):
    plt.plot(points[i][0],points[i][2],marker='o', markersize=3, color="red")
plt.figure(1)
plt.plot(times,all_time_distance)
plt.plot(times,means_distance)
for i in range(len(points)):
    plt.plot(points[i][0],points[i][1],marker='o', markersize=3, color="red")
plt.show()
```
