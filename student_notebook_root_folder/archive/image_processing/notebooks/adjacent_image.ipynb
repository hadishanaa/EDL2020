```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

#### First image

```python
inputPath = '../videos/recording1.avi'
cap = cv2.VideoCapture(inputPath)
ret, frame = cap.read()
plt.imshow(frame[...,::-1])
print("type {}, shape {}".format(type(frame), frame.shape))
```

```python
inputPath2 = '../videos/fd1.avi'
cap2 = cv2.VideoCapture(inputPath2)
ret2, frame2 = cap2.read()
plt.imshow(frame2[...,::-1])
print("type {}, shape {}".format(type(frame2), frame2.shape))
```

```python
#This puts each image array side by side
new_img = np.concatenate([frame, frame2], axis = 1)
plt.imshow(new_img[...,::-1])
print("type {}, shape {}".format(type(frame2), frame2.shape))
```
